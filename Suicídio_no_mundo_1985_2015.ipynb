<h3>Origem do Setembro Amarelo</h3><br>

O Setembro Amarelo teve sua origem nos EUA, quando um jovem chamado Mike Emme, de 17 anos, cometeu suicídio, em 1994.

Em 1994, o americano Mike Emme, filho do casal Dale Emme e Darlene Emme, se suicidou com apenas 17 anos. Mike era conhecido por sua personalidade caridosa e por saber muito sobre mecânica. Sozinho, o garoto conseguiu restaurar um Mustang 68 e pintou o carro todo de amarelo.

Seus pais e amigos não perceberam que o jovem tinha sérios problemas psicológicos e não conseguiram evitar sua morte.

No dia do velório, foi feita uma cesta com muitos cartões decorados com fitas amarelas. Dentro deles tinha a mensagem "Se você precisar, peça ajuda.". A iniciativa foi o estopim para um movimento importante de prevenção ao suicídio, pois os cartões chegaram realmente às mãos de pessoas que precisavam de apoio.

Em consequência dessa triste história, foi escolhido como símbolo da luta contra o suicídio, o laço amarelo.

```python
# função para calcular o total de casos por ano
def total_por_ano(frame, new_col_name, list_of_cols_to_sum):
    frame[new_col_name] = frame[list_of_cols_to_sum].astype(float).sum(axis=1)
    return(frame)

# criar uma tabela dinâmica
tabela = pd.pivot_table(df, values='suicides_no', index=['year'], columns=['age'])

# reordenar as tabelas para deixar em ordem crescente
column_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years',]
tabela = tabela.reindex(column_order, axis=1)


# verificando o total de suicídios por ano no mundo
total_por_ano(tabela, 'total', ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'])

# ver as 5 primeiras entradas da tabela
tabela.head()
```

```python
df_brasil = df[df.country == "Brazil"].copy()

# ver o formato do novo DataFrame
print("Formato do DataFrame: ", df_brasil.shape)

# exibir as 5 primeiras entradas do novo DataFrame
df_brasil.head(10)
```

```python
#tá certo
df_brasil.groupby(by=["year"]).sum()
```

```python
def total_por_ano_brasil(frame, new_col_name, list_of_cols_to_sum): 
  frame[new_col_name] = frame[list_of_cols_to_sum].astype(float).sum(axis=1) 
  return(frame)

#criar uma tabela dinâmica
tabela_brasil = pd.pivot_table(df_brasil, values=['suicides_no'], index=['year'], columns=['age'])

#reordenar as tabelas para deixar em ordem crescente
column_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years',] 
tabela_brasil = tabela_brasil.reindex(column_order, axis=1)
#verificando o total de suicídios por ano no mundo
total_por_ano_brasil(tabela_brasil, 'total_brasil', ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'])

#ver as 5 primeiras entradas da tabela
tabela_brasil.head()
```

```python
tabela_brasil['total_brasil'].plot(figsize=(18,10));
plt.title("Total de suicídios por ano");
plt.legend();
```

```python
tabela_brasil['total_brasil'].plot(figsize=(18,10));
tabela['total'].plot(figsize=(18,10));
plt.title("Total de suicídios por ano Mundo x Brasil");
plt.legend();
```
